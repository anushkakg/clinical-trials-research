# Fields Analysis


```python
import httpx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
API_URL = "https://clinicaltrials.gov/api/v2/studies?pageSize=1000"
```

```python
response_json = httpx.get(API_URL).json()
df = pd.json_normalize(response_json["studies"])

while response_json.get("nextPageToken", False):
    next_page_token = response_json["nextPageToken"]
    response_json = httpx.get(API_URL + f"&pageToken={next_page_token}").json()
    df = pd.concat([df, pd.json_normalize(response_json["studies"])], ignore_index=True)
```

```python
df.columns = [col.split(".")[-1] for col in df.columns]
```
