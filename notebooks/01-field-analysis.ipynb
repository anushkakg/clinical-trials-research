# Fields Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Database

```python
DB_URL = "sqlite:///../clinicaltrials.db"
df = pd.read_sql_table("enumfields", DB_URL)
df[["piece", "missingStudiesCount", "uniqueValuesCount"]]
```

## Visualize Field Statistics

```python
plt.figure(figsize=(14, 8))
plt.bar(df["piece"], df["uniqueValuesCount"])
plt.xlabel("Field")
plt.ylabel("Unique Values")
plt.xticks(rotation=90)
plt.title("Unique Values for Each ENUM field")
plt.tight_layout()
plt.show()
```
